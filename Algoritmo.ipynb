```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Datos ficticios: dias y consumo de aguas en litros.
#Dia: 1, 2, 3
#Consumo (litros)
dias = np.array([1, 2, 3]).reshape(-1, 1)
consumo = np.array([120, 125, 145])

#Creacion de modelo a entrenar
modelo = LinearRegression()
modelo.fit(dias, consumo)

#Coeficientes
pendiente = modelo.coef_[0]
interseccion = modelo.intercept_

print("Modelo entrenado:")
print(f" - Pendiente: {pendiente}")
print(f" - Interseccion: {interseccion}")
```

```python
dia_prediccion = np.array([[4]])
consumo_predicho = modelo.predict(dia_prediccion)
print(f"\n Prediccion para el dia 4: {consumo_predicho[0]:.2f} litros")
```

```python
#Grafico mostrado
plt.scatter(dias, consumo, color='blue', label='Datos reales')
plt.plot(dias, modelo.predict(dias), color='red', label='Linea de regresion')
plt.scatter(dia_prediccion, consumo_predicho, color='green', label='Prediccion (dia 4)')
plt.title('Consumo de Agua por Dia')
plt.xlabel('Dia')
plt.ylabel('Consumo (litros)')
plt.legend()
plt.grid(True)
plt.show()
```
